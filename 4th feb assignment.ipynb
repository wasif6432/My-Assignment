 ans 1) # Given list of tuples
list_of_tuples = [('Sachin Tendulkar', 34357), 
                  ('Ricky Ponting', 27483), 
                  ('Jack Kallis', 25534), 
                  ('Virat Kohli', 24936)]

# Sorting the list of tuples based on the second element (integer value) using a lambda function
sorted_list = sorted(list_of_tuples, key=lambda x: x[1])

# Printing the sorted list
print("Sorted list of tuples based on integer value:")
for item in sorted_list:
    print(item)

#output of this program:

Sorted list of tuples based on integer value:
('Virat Kohli', 24936)
('Jack Kallis', 25534)
('Ricky Ponting', 27483)
('Sachin Tendulkar', 34357)

# Printing the squared numbers
print("Squares of all the numbers in the list:")

from functools import reduce

# Create a list containing numbers from 1 to 25
numbers = list(range(1, 26))

# Compute the product of the numbers using reduce() function
product = reduce(lambda x, y: x * y, numbers)

# Printing the product
print("Product of numbers from 1 to 25:")
print(product)

# Given list of numbers
numbers = [2, 3, 6, 9, 27, 60, 90, 120, 55, 46]

# Using filter() with lambda to filter numbers divisible by both 2 and 3
filtered_numbers = filter(lambda x: x % 2 == 0 and x % 3 == 0, numbers)

# Printing the filtered numbers
print("Numbers divisible by both 2 and 3:")
print(list(filtered_numbers))

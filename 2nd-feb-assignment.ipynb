1)For Loop:
A for loop is ideal when you know beforehand how many times you want to repeat a specific block of code. It's typically used for iterating over a sequence of items like a list, array, or a range of numbers.

Example: Calculating the sum of numbers from 1 to 10.

sum = 0
for i in range(1, 11):  # The range function generates numbers from 1 to 10.
    sum += i
print("The sum is:", sum)

 While Loop:
A while loop is used when you want to repeat a block of code as long as a certain condition is true. You might not know in advance how many iterations you need.

Example: Repeatedly asking the user for input until they enter a valid number.

valid_input = False
while not valid_input:
    user_input = input("Enter a number: ")
    if user_input.isdigit():
        valid_input = True
    else:
        print("Invalid input. Please enter a valid number.")
print("You entered a valid number:", user_input)

# Using a for loop to calculate the cubes and find numbers divisible by 4 or 5
cubes_list = []

for number in range(1, 101):
    cube = number ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_list.append(number)

print("Using a for loop:")
print("Numbers whose cubes are divisible by 4 or 5:")
print(cubes_list)
